```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv(r'C:\Users\deeks\Downloads\sales_data.csv.csv')

# Step 2: Convert order_date to datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Step 3: Filter for 2022 only
df_2022 = df[df['order_date'].dt.year == 2022]

# Step 4: Group by category and month
df_agg = df_2022.groupby([
    'category',
    pd.Grouper(key='order_date', freq='M')
])['qty_ordered'].sum().reset_index()

# Step 5: Plot monthly sales trends
plt.figure(figsize=(14, 6))
sns.lineplot(
    data=df_agg,
    x='order_date',
    y='qty_ordered',
    hue='category',
    marker='o'
)
plt.title('Monthly Sales Trends by Category - 2022')
plt.xlabel('Month')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Ensure order_date is in datetime format
df_agg['order_date'] = pd.to_datetime(df_agg['order_date'])

# Filter for first month (January) and last month (December)
first_month = df_agg[df_agg['order_date'].dt.month == 1]
last_month = df_agg[df_agg['order_date'].dt.month == 12]

# Check if both dataframes are empty (i.e., no data for January or December)
if first_month.empty or last_month.empty:
    print("No data available for January or December.")
else:
    # Merge the dataframes based on 'category'
    growth_df = pd.merge(
        first_month[['category', 'qty_ordered']],
        last_month[['category', 'qty_ordered']],
        on='category',
        suffixes=('_jan', '_dec')
    )
    
    # Check if the merge resulted in an empty dataframe
    if growth_df.empty:
        print("No matching categories found between January and December.")
    else:
        # Calculate absolute and percentage growth
        growth_df['abs_growth'] = growth_df['qty_ordered_dec'] - growth_df['qty_ordered_jan']
        
        # Avoid division by zero or NaN values
        growth_df['pct_growth'] = (growth_df['abs_growth'] / growth_df['qty_ordered_jan'].replace(0, pd.NA)) * 100
        
        # Handle NaN values after division
        growth_df = growth_df.dropna(subset=['pct_growth'])

        # Sort by percentage growth to find the top category
        if not growth_df.empty:
            top_growth_category = growth_df.sort_values('pct_growth', ascending=False).iloc[0]
            print("Top growth category:", top_growth_category)
        else:
            print("No valid growth data available.")
```

```python
import pandas as pd

# Sample data (replace with your actual data)
data = {
    'customer_id': [1, 2, 3, 4],
    'after_discount': [200, 300, 50, 600],
    'cogs': [100, 150, 20, 100]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate net profit (after_discount - cogs)
df['net_profit'] = df['after_discount'] - df['cogs']

# Define the categorize_profit function
def categorize_profit(net_profit):
    if net_profit < 100:
        return 'Low'
    elif 100 <= net_profit <= 500:
        return 'Medium'
    else:
        return 'High'

# Apply the categorize_profit function to the net_profit column
df['profit_segment'] = df['net_profit'].apply(categorize_profit)

# Display the result
print(df[['customer_id', 'after_discount', 'cogs', 'net_profit', 'profit_segment']])
```

```python
import matplotlib.pyplot as plt

# Calculate the distribution of customers across segments
segment_counts = df['profit_segment'].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'orange', 'lightgreen'])
plt.title('Customer Distribution by Profit Segment')
plt.show()
```
